```python
import bson
import gzip

class open_bson:
    def __init__(self, path):
        self.path = path
    
    def __enter__(self):
        open_file_func = open
        if ".gz" == fileName[-3:]:
            open_file_func = gzip.open
        self.f = open_file_func(fileName, 'rb')
        return bson.decode_file_iter(self.f)
    
    def __exit__(self, *exc_info):
        self.f.close()
        del self.f
```

```python
known_influencers = [('Henrique Bastos', '‏@henriquebastos', 14227855),
('Michael Kennedy', '@mkennedy', 12953262),
('Talk Python Podcast', '@TalkPython', 3098427092),
('Russel Winder', '@russel_winder', 238541848),
('Richard Jones', '‏@r1chardj0n3s', 14379613),
('Mikko Ohtamaa', '‏@moo9000', 50703014),
('Tibor Sziráki', '@openerphu', 173394343),
('Data Science Renee', '‏@BecomingDataSci', 2343198944),
('steve_piercy', '@steve_piercy', 93713345)]
known_influencers_dict = {x[2]:x for x in known_influencers}
known_influencers_set = {x[2] for x in known_influencers}
def get_known_influencers(df):
    return [x for x in df['id'] if x in known_influencers_set]
```

```python
account_object_field = 'account_object'
account_fields = ['contributors_enabled','created_at','description','entities',
                        'favourites_count','followers_count','friends_count',
                        'id','id_str','is_translator','lang','listed_count','location','name',
                        'profile_background_tile','protected','screen_name',
                        'show_all_inline_media','statuses_count','url','verified']
def parse_account_object(d):
    res = []
    for name in account_fields:
        if name in d:
            res.append(d[name])
        else:
            res.append(np.nan)
    return res
```

```python
fileName = "C:/Users/dkorolev/Downloads/twitter-analysis_last_last/Accounts.bson"
accounts = []
last_status_update = []
with open_bson(fileName) as f:
    for dict_data in f:
        if account_object_field in dict_data and len(dict_data[account_object_field]) > 1:
            if 'status' in dict_data[account_object_field]:
                last_status_update.append(dict_data[account_object_field]['status']['created_at'])
            else:
                last_status_update.append('')
            accounts.append(parse_account_object(dict_data[account_object_field]))
```

```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import MinMaxScaler
np.set_printoptions(suppress=True)
scaler = MinMaxScaler()
all_X = scaler.fit_transform(accounts_df[fields_for_clustering])
all_tsne_model = TSNE(n_components=2, random_state=0)
all_tsne = all_tsne_model.fit_transform(all_X)
```

```python
with open('pycharm_followers_tsne.pkl', 'wb') as f:
    pickle.dump(all_tsne, f)
```

```python
influencers_ids = [x for x in range(accounts_df.shape[0]) if accounts_df.iloc[x]['id'] in known_influencers_set]
```

```python
plt.plot(all_tsne[:,0], all_tsne[:,1], 'o')
plt.plot(all_tsne[influencers_ids,0], all_tsne[influencers_ids,1], 'or')
```

```python
from scipy.cluster import hierarchy
from scipy.spatial.distance import pdist
dist = pdist(all_X, 'euclidean')
```

```python
Z = hierarchy.linkage(all_X, method='ward')
```

```python
with open('pycharm_followers_ward_linkage.pkl', 'wb') as f:
    pickle.dump(Z, f)
```

```python
def hierarchy_draw(Z, labels, level):
    plt.figure()
    hierarchy.dendrogram(Z, labels=labels, color_threshold=level, leaf_font_size=5, count_sort=True)
    plt.show()
```

```python
hierarchy_draw(Z, np.arange(all_X.shape[0]), 0.35)
```

```python
def plot_clustering(X_2D, labels, axes, title=None):
    clusters_count = len(set(labels))
    for i in range(1, clusters_count + 1):
        X_2D_part = X_2D[labels == i]
        axes.plot(X_2D_part[:, 0], X_2D_part[:, 1], 'o', color=plt.cm.spectral(i / clusters_count))
    axes.plot(X_2D[influencers_ids,0], X_2D[influencers_ids,1], 'or')
    if title is not None:
        axes.set_title(title, size=17)
```

```python
n_clusters = 10
fig, axes = plt.subplots(nrows=(n_clusters-1), ncols=1, figsize=(10, 5*(n_clusters - 1)))
for i in range(2,n_clusters + 1):
    labels = hierarchy.fcluster(Z, t=i, criterion='maxclust')
    plot_clustering(all_tsne, labels, axes[i-2], "%d clusters" % i)
```
